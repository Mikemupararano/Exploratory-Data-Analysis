```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn  as sns
from sklearn.datasets import load_iris
```

```python
#Load the dataset
iris=load_iris()
data= pd.DataFrame(data= iris.data, columns=iris.feature_names)
data.head()
```

```python
data['species']=iris.target_names[iris.target]
data.head()
```

```python
#Return the shape of data frame and here there 150 observations and 5 variables in total (n_observations, n_varibales)
data.shape
```

```python
print("Data shape: ", data.shape)
print("Features: ", data.columns [:-1])
```

```python
#Check for misisng values
data.isnull().sum() #no missing values in the dataset.
```

```python
data.describe()
```

```python
#Visualisation using histogram. Univariate analysis
data.hist()
plt.tight_layout()
```

```python
#Plot boxplot to find if there are outliers and there are 4 outliers here.
data.boxplot()
```

```python
#Bivariate analysis
sns.pairplot(data)
```

```python
sns.pairplot(data, hue='species')
```

```python
#Quantify correlations using heatmap
corr_matrix=data
data.head()
```

```python
#Exclude strings and last columns
corr_matrix= data.iloc[:, :-1].corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot= True, cmap ='coolwarm')
# Closer to 1 =strong positive correlation.
#Closer to -1 =strong strong negative correaltion.
#Close to 0 means no correaltions.
#Sitting around 0.5/-0.5 indicates pretty moderate positive/negative correaltion.
```
